```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
loan_data=pd.read_csv('/content/Loan_Status_Prediction.csv')
```

```python
loan_data.head()
```

```python
loan_data.shape
```

```python
loan_data.isnull().sum()
```

```python
loan_data=loan_data.dropna()
```

```python
loan_data['Loan_Status'].value_counts()
```

```python
loan_data.replace({"Loan_Status":{"N":0,"Y":1}},inplace=True)
```

```python
loan_data=loan_data.replace(to_replace='3+',value=4)
```

```python
# covert data to numric values
loan_data.replace({'Married':{'Yes':1,"No":0},
                   "Gender":{"Female":0,"Male":1},
                   "Education":{"Graduate":1,"Not Graduate":0},
                   "Self_Employed":{"Yes":1,"No":0},
                   "Property_Area":{"Rural":0,"Semiurban":1,"Urban":2}},inplace=True)
```

```python
loan_data
```

```python
X=loan_data.drop(['Loan_Status','Loan_ID'],axis=1)
Y=loan_data['Loan_Status']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2,stratify=Y)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_predactions=classifier.predict(X_train)
X_train_predaction_score=accuracy_score(Y_train,X_train_predactions)
```

```python
X_train_predaction_score
```

```python
X_test_predactions=classifier.predict(X_test)
X_test_predaction_score=accuracy_score(Y_test,X_test_predactions)
```
